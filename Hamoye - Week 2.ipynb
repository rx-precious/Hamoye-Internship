```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data1 = data.drop(['lights', 'date'], axis=1)
```

```python
data1.head()
```

```python
data1.describe()
```

```python
#Data Normalization usin MinMax Scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_data = scaler.fit_transform(data1)
normalised_data = pd.DataFrame(normalised_data, columns= data1.columns )
normalised_data.head()
```

```python
#Splitting the data into features and labels
label = 'Appliances'
data_features = normalised_data.drop(label, axis = 1 )
data_label = normalised_data[label]
data_features.head()
```

```python
data_label.head()
```

```python
#Splitting dataset
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(data_features, data_label, test_size = 0.3, random_state = 42)
X_train.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

lr = LinearRegression()
ridge = Ridge(alpha= 0.4)
lasso = Lasso(alpha=0.001)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
#12 Finding the relationship between T2 and T6

xt =pd.DataFrame(X_train['T2'])
yt = X_train['T6']
xtx= pd.DataFrame(X_test['T2'])
ytx = X_test['T6']

lr.fit(xt,yt)
p  = lr.predict(xtx)
r2 = round(lr.score(xtx, ytx),2)
print('The r2/COD is {}'.format(r2))


```

```python
def fit(model, X_t, y_t):
    model.fit(X_t, y_t)
    return model
```

```python
lrr = fit(lr, X_train, y_train)
r = fit(ridge, X_train, y_train)
l = fit(lasso, X_train, y_train)

pred_lr = lrr.predict(X_test)
pred_r = r.predict(X_test)
pred_l = l.predict(X_test)
```

```python

r2 = r2_score(y_test,pred_lr).round(2)
print('The r2/COD is {}'.format(r2))

mae = mean_absolute_error(y_test, pred_lr).round(2)
print('The MAE is {}'.format(mae))
```

```python
#for sum of residual squares
rss = np.sum((y_test - pred_lr)**2).round(2)
print('The RSS is {}'.format(rss))
```

```python
msr = np.sqrt( mean_squared_error(y_test, pred_lr)).round(3)
print('The MSR with Linear Regression is {}'.format(msr))
```

```python
msrlasso = np.sqrt( mean_squared_error(y_test, pred_l)).round(3)
print('The MSR with LASSO Regression is {}'.format(msrlasso))
```

```python
#compare the effects of regularization
def get_weights(model, features, col_name):
    weights = pd.Series(model.coef_, features.columns).sort_values()
    weights = pd.DataFrame(weights).reset_index()
    weights.columns = ['Features', col_name]
    weights[col_name].round(3)
    return weights
```

```python
linear = get_weights(lr, X_train, 'Linear_Model Weights')
ridge = get_weights(r, X_train, 'Ridge_Model Weights')
Lasso = get_weights(l, X_train, 'LASSO_Model Weights')

Lasso
```
